```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = pd.read_csv("customer_shopping_behavior.csv")
file.head()
```

```python
file.info()
```

```python
file.describe(include='all')
```

```python
file.isnull().sum()
```

```python
file['Review Rating'] = file['Review Rating'].fillna(file.groupby('Category')['Review Rating'].transform('median'))
```

```python
file.info()
```

```python
#verifying whether Promo Code Used and Discount Applied are truly identical
pd.crosstab(file['Promo Code Used'],file['Discount Applied'])
```

```python
#drop column Promo Code Used
file.drop(columns = ['Promo Code Used'],inplace=True)
```

```python
file.info()
```

```python
#change column name writing style
file.columns = file.columns.str.lower()
file.columns = file.columns.str.replace(' ', '_')
```

```python
file.info()
```

```python
file.columns = file.columns.str.replace('(usd)','usd')
```

```python
file.info()
```

```python
#create a column age_group
labels = ['young adult','adult','middle aged','senior']
file['age_group'] = pd.qcut(file['age'],q = 4, labels = labels)
```

```python
file.head()
```

```python
file['frequency_of_purchases'].unique()
```

```python
#create new column purchase frequency days
frequency_mapping =     {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly': 30,
    'Every 3 Months': 90
                        }
```

```python
file['purchase_frequency_days'] = file['frequency_of_purchases'].map(frequency_mapping)
```

```python
file['purchase_frequency_days']
```

```python
file.info()
```

```python
file_copy = file.copy()
```

```python
#delete column frequency_of_purchases
file_copy = file_copy.drop('frequency_of_purchases',axis = 1)
```
